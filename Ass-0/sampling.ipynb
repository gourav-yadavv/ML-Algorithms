A sample of size 10 is randomly selected (without replacement) from the original data.

```python
import pandas as pd

# Loading the dataset
emnist_ds = pd.read_csv("emnist-byclass-test.csv", header=None)

# Define the number of rows to sample
sample_size = 10

# Randomly sample the rows from the dataset
sampled_ds = emnist_ds.sample(n=sample_size)

# Save the sampled dataset to a new CSV file
# sampled_df.to_csv("emnist-sample.csv", header=False, index=False)
sampled_ds
```

I randomly select 1% of the data (without replacement) and display the selected samples. The random_state argument of the function specifies the seed value of the random number generator.

```python
sample = emnist_ds.sample(frac=0.01, random_state=1)
sample
```

Finally, I performed a sampling with replacement to create a sample whose size is equal to 1% of the entire data.

```python
sample = emnist_ds.sample(frac=0.01, replace=True, random_state=1)
sample
```
